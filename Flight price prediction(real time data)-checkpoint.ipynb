# Flight price prediction
- Anyone who has booked a flight ticket knows how unexpectedly the prices vary. The cheapest 
  available ticket on a given flight gets more and less expensive over time. This usually happens as 
  an attempt to maximize revenue based on -
1. Time of purchase patterns (making sure last-minute purchases are expensive)
2. Keeping the flight as full as they want it (raising prices on a flight which is filling up in order 
   to reduce sales and hold back inventory for those expensive last-minute expensive purchases) So, you have to work on a     
   project where you collect data of flight fares with other features and work to make a model to predict fares of flights.

```python
## Divide data set into features and label
y = df['Price']
x = df.drop(columns=['Price'], axis=1)
```

##### Finding best Random_ state

```python
from sklearn.ensemble import RandomForestRegressor
maxr2score = 0
maxRS = 0

for i in range (1, 150):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = i)
    RF = RandomForestRegressor()
    RF.fit(x_train, y_train)
    predrf = RF.predict(x_test)
    r2 = r2_score(y_test, predrf)
    if r2>maxr2score:
        maxr2score = r2
        maxRS = i
print ('Max R2 score obtained is', maxr2score, 'on Random State', maxRS)
```

##### Creating train test split

```python
# Split the data into train and test. Model will be built on training data and tested on test data.
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=maxRS)
y_train.head()
```

##### Hyper parameter tuning

```python
params = {'n_estimators': range(50,200,10),
               'max_features': ['auto', 'sqrt'],
               'max_depth': range(5,30,6),
               'min_samples_split': [2, 5, 10, 15, 100],
               'min_samples_leaf': [1, 2, 5, 10]}
```

```python
from sklearn.model_selection import RandomizedSearchCV
rf = RandomizedSearchCV(RandomForestRegressor(),params,scoring='neg_mean_squared_error',n_iter = 10,cv=5, n_jobs= -1,random_state=42)
rf.fit(x_train,y_train)
```

```python
rf.best_params_
```

```python
pred = rf.predict(x_test)
print(r2_score(y_test,pred))
```
